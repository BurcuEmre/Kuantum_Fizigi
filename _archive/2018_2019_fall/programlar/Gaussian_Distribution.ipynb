```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def gaussk(k):
    return np.exp(-alfa*(k-k0)**2/2)

def gaussx(x):
    return np.exp(-x**2/2/alfa)

def gauss2t(x, t):
    return 2*np.pi*np.exp(-alfa*(x-v_g*t)**2/(alfa**2 + 4 * beta**2 * t**2))/np.sqrt(alfa**2 + 4 * beta**2 * t**2)
```

```python
alfa = 4.
k0 = 1.
x0 = 1.

k = np.linspace(-5+k0, 5+k0, 250)
x = np.linspace(-5+x0, 5+x0, 250)

kwidth = 2/np.sqrt(alfa)
kminus = k0 - kwidth/2
kplus  = k0 + kwidth/2

xwidth = 2*np.sqrt(alfa)
xminus = x0 - xwidth/2
xplus  = x0 + xwidth/2


plt.figure(figsize=(9, 6))

plt.plot(k, gaussk(k)**2, label=u"k dağılımı")
plt.plot([kminus, kminus, kplus, kplus], [0, gaussk(kminus)**2, gaussk(kplus)**2, 0], label=u"k genişlik")

plt.plot(x, gaussx(x-x0)**2, label=u"x dağılımı")
plt.plot([xminus, xminus, xplus, xplus], [0, gaussx(xminus-x0)**2, gaussx(xplus-x0)**2, 0], label=u"x genişlik")

plt.grid()
plt.xlabel("x, k", fontsize="18")
plt.ylabel("g(x), g(k)", fontsize="18")
plt.legend(loc="best")
plt.show()
```

### $\beta = 0$ iken dalga paketi dağılmadan $v_g$ grup hızıyla ilerler

```python
alfa = 4.0
beta = 0.0
v_g = 3.0

k0 = 1.
x0 = 1.

x = np.linspace(-20, 140, 500)

for t in range (0, 31, 5):
    plt.plot(x, gauss2t(x, t), label="t = %s s"%t)

plt.legend(loc="best")
plt.xlabel("$x$", fontsize=18)
plt.ylabel("$\psi(x,t)$", fontsize=18)
plt.show()
```

### $\beta \neq 0$ iken dalga paketi dağılımlı ortamda yayılarak ilerler.

```python
alfa = 4.0
beta = 0.5
v_g = 3.0

k0 = 1.
x0 = 1.

x = np.linspace(-20, 140, 500)

for t in range (0, 31, 5):
    plt.plot(x, gauss2t(x, t), label="t = %s s"%t)

plt.legend(loc="best")
plt.xlabel("$x$", fontsize=18)
plt.ylabel("$\psi(x,t)$", fontsize=18)
plt.show()
```
